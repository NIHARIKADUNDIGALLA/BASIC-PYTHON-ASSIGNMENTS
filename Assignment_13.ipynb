```python
#1)Can you create a programme or function that employs both positive and negative indexing? Is there any repercussion 
#if you do so?

#Ans)
##positive indexing
def fun1(*args):
        for i in range(len(args)):
            print("index",i,"for an element",args[i])

fun1("niha",2,3)

##negative indexing
def fun2(*args):
        for i in range(len(args)-1,-1,-1):
            print("index",i-3,"for an element",args[i])
            
fun2(3,4,[3,4])
```
